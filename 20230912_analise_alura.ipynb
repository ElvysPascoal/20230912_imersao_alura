```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Visualize as primeiras linhas do DataFrame
print(df.head())
```

```python
# Verifique as informações gerais do DataFrame
print(df.info())
```

```python
# Resumo estatístico das colunas numéricas
print(df.describe())
```

```python
# Verifique a contagem de valores únicos em cada coluna
print(df.nunique())
```

```python
# Histograma de uma coluna numérica (por exemplo, 'actual_price')
plt.figure(figsize=(8, 6))
sns.histplot(df['actual_price'], bins=20, kde=True)
plt.title('Histograma de Preços Atuais')
plt.xlabel('Preço Atual')
plt.ylabel('Frequência')
plt.show()
```

```python
# Gráfico de dispersão (scatter plot) entre duas colunas numéricas (por exemplo, 'actual_price' vs 'product_price')
plt.figure(figsize=(8, 6))
sns.scatterplot(x='actual_price', y='product_price', data=df)
plt.title('Gráfico de Dispersão entre Preço Atual e Preço do Produto')
plt.xlabel('Preço Atual')
plt.ylabel('Preço do Produto')
plt.show()
```

```python
# Correlação entre colunas numéricas
correlation_matrix = df.corr()
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Matriz de Correlação')
plt.show()
```

```python
# Contagem de Valores Únicos na Coluna 'category':
category_counts = df['category'].value_counts()
print(category_counts)
```

```python
# Gráfico de Barra de Rating:
plt.figure(figsize=(8, 6))
sns.histplot(df['rating'], bins=5, kde=True)
plt.title('Distribuição de Classificações de Produtos')
plt.xlabel('Classificação')
plt.ylabel('Frequência')
plt.show()
```

```python
# Gráfico de Dispersão entre Preço Atual e Avaliação:
plt.figure(figsize=(8, 6))
sns.scatterplot(x='actual_price', y='rating', data=df)
plt.title('Gráfico de Dispersão entre Preço Atual e Avaliação')
plt.xlabel('Preço Atual')
plt.ylabel('Avaliação')
plt.show()
```

```python
# Análise de Valores Ausentes:
missing_values = df.isnull().sum()
print(missing_values)
```

```python
# Contagem de Produtos com Desconto:
discounted_products = df[df['percent'] > 0]
discounted_count = discounted_products.shape[0]
print(f"Total de produtos com desconto: {discounted_count}")
```
